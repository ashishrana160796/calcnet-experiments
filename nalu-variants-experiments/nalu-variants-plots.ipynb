```python
# For rmse based mlp model.
# special turn-around of loading custom objects is required.

def rmse(y_true, y_pred):
    return K.sqrt(K.mean(K.square(y_pred - y_true)))

m_add = load_model('mlp_rmse_add.h5', custom_objects={'rmse': rmse})
m_add.evaluate(tstx_add, tsty_add)
```
